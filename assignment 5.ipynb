```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Load your data (change the folder path to your dataset folder)
folder_path = r"C:\Users\Lenovo\CSIR NIO\track data\1194\excel"  # Replace with the path to your folder
all_data = []

# Read and filter CSV files
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        data = pd.read_csv(file_path)
        data_filtered = data[(data['confidence code'] == 4) & 
                             (data['photon height'] >= -79) & 
                             (data['photon height'] <= -75)]
        all_data.append(data_filtered)

# Combine all data into one DataFrame
all_data = [df for df in all_data if not df.empty]  # Exclude empty DataFrames
combined_data = pd.concat(all_data, ignore_index=True)

# Sample a fraction of the data to avoid memory issues during clustering
sampled_data = combined_data.sample(frac=0.1, random_state=42)

# Extract latitude, longitude, and photon height
X_sampled = sampled_data[['latitude', 'longitude', 'photon height']]

# Standardize the data
scaler = StandardScaler()
X_scaled_sampled = scaler.fit_transform(X_sampled)

# Compute k-distances
k = 4  # Number of neighbors for k-distance plot
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X_scaled_sampled)
distances, indices = neighbors_fit.kneighbors(X_scaled_sampled)

# Sort distances to get the k-distance
sorted_distances = np.sort(distances[:, k-1], axis=0)  # Distance to the kth nearest neighbor
average_distance = np.mean(sorted_distances)  # Calculate the average k-distance

# Determine potential `eps` by finding the point where the decrease in distance slows
change_in_distances = np.diff(sorted_distances)
eps_candidate = sorted_distances[np.argmax(change_in_distances)]  # Select where the decrease is sharpest

# Print output for eps and average distance
print(f"Average k-distance: {average_distance:.4f}")
print(f"Suggested eps value: {eps_candidate:.4f}")

# Optionally, save the calculated distances for further analysis
# np.savetxt("k_distances.csv", sorted_distances, delimiter=",")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Define the folder path containing the CSV files
folder_path = r"C:\Users\Lenovo\CSIR NIO\track data\1194\excel"  # Replace with the path to your folder

# Initialize an empty list to store data from all CSV files
all_data = []

# Loop through all CSV files in the folder and read them
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Read the current CSV file
        data = pd.read_csv(file_path)
        # Filter to include only rows with confidence code 4 and photon height between -75 and -79
        data_filtered = data[(data['confidence code'] == 4) & 
                             (data['photon height'] >= -79) & 
                             (data['photon height'] <= -75)]
        all_data.append(data_filtered)

# Combine all data into a single DataFrame
all_data = [df for df in all_data if not df.empty]  # Exclude empty DataFrames
combined_data = pd.concat(all_data, ignore_index=True)

# Extract features for clustering
X = combined_data[['latitude', 'longitude', 'photon height']]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN with the suggested eps and min_samples values
eps_value = 0.3198  # From the output
min_samples_value = 1000  # Adjust this if necessary

dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
labels = dbscan.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
combined_data['cluster'] = labels

# Plot 1: Latitude vs Longitude vs Photon Height (3D scatter plot)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(combined_data['latitude'], combined_data['longitude'], combined_data['photon height'], 
                     c=combined_data['cluster'], cmap='tab20', marker='o')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Photon Height')
ax.set_title('3D Scatter Plot: Latitude vs Longitude vs Photon Height')
fig.colorbar(scatter, ax=ax, label='Cluster ID')

# Plot 2: Latitude vs Photon Height
plt.figure(figsize=(10, 6))
plt.scatter(combined_data['latitude'], combined_data['photon height'], c=combined_data['cluster'], cmap='tab20', marker='o')
plt.xlabel('Latitude')
plt.ylabel('Photon Height')
plt.title('Latitude vs Photon Height')
plt.colorbar(label='Cluster ID')

# Plot 3: Longitude vs Photon Height
plt.figure(figsize=(10, 6))
plt.scatter(combined_data['longitude'], combined_data['photon height'], c=combined_data['cluster'], cmap='tab20', marker='o')
plt.xlabel('Longitude')
plt.ylabel('Photon Height')
plt.title('Longitude vs Photon Height')
plt.colorbar(label='Cluster ID')

plt.tight_layout()
plt.show()
```
